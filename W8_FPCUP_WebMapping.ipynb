```python
# call the leafmap folium
import leafmap.foliumap as leafmap
```

```python
# Import geopandas to explore the vector data
import geopandas as gpd
```

```python
# Connect to your Google drive to read your geospatial data
from google.colab import drive
drive.mount('/content/gdrive')
```

```python
# Read the Data file path
town_fd = "/content/gdrive/MyDrive/FPCUP_2ndP_ToT/data/town.geojson"
```

```python
# Read the vector data with geopandas
town = gpd.read_file(town_fd)
```

```python
# Explore the table attribute of the data
town
```

```python
town.plot()
```

```python
muni_fd = '/content/gdrive/MyDrive/FPCUP_2ndP_ToT/data/municipality.geojson'
```

```python
muni = gpd.read_file(muni_fd)
```

```python
muni
```

```python
muni.plot()
```

```python
# Define the visualization style
my_style = {
    "stroke": True,
    "color": "red",
    "weight": 2,
    "opacity": 1,
    "fill": True,
    "fillColor": "#0000ff",
    "fillOpacity": 0.1,
}
```

```python
# Add the layers to Leaflet folium
my_map = leafmap.Map(center=[-20.90556,55.60747], zoom=10)
my_map.add_basemap("HYBRID")
my_map.add_geojson(town_fd, layer_name='Cities')
my_map.add_geojson(muni_fd,layer_name = 'Municipalities', style = my_style)
my_map
```

```python
# Explore the HTML structure
html = my_map.to_html()
print(html)
```

```python
# Save it as HTML for Webmapp deployment
my_map.to_html("my_webmap.html")
```
